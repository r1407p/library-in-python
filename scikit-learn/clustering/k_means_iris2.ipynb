```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
print(iris.feature_names)
X=iris.data
X=X[:,2:4]#use petal length and width
print(X)
```

```python
KM_model=KMeans(n_clusters=3,init='random',random_state=12)
KM_model.fit(X)
```

```python
KM_model.predict(X)
ans = pd.DataFrame(KM_model.labels_)
print(ans)
```

```python
plt.figure(figsize= (10,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c=iris.target)
plt.title("Answer Of classification")
plt.subplot(1,2,2)
plt.scatter(X[:,0],X[:,1],c=KM_model.predict(X))
plt.title("K Means Classificaion")
```
